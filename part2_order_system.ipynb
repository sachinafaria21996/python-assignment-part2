```python
# TASK 2: Cart Operations

cart=[]                                                                # Initialize an empty cart to store items added by the user

def add_to_cart(item_name, quantity):                                  # Function to add items to the cart with necessary checks and updates

    # Sub Task 1. Check if item exists in menu

    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in the menu.")
        return
    
    # Sub Task 2. Check if item present in menu is available

    if menu[item_name]["available"] == False:
        print(f"Error: '{item_name}' is currently unavailable.")
        return

    # Sub Task 3. Check if item is already present in cart
    for add in cart:                                            # For Loop Iteration through cart to find if an item is already added
        if add["item"] == item_name:                            # Check if added item name is same as an item already present in cart
            add["quantity"] += quantity                         # If item is already present, update the quantity by adding the new quantity to existing quantity    
            print(f"Updated '{item_name}' quantity to {add['quantity']}.")
            return
    
    # Sub Task 4. If item is not present in cart, add it as a new entry

    new_entry = {
        "item": item_name,                                      # Store the item name in the new entry
        "quantity": quantity,                                   # Store the quantity of the item in the new entry 
        "price": menu[item_name]["price"]                       # Store the price of the item from the menu in the new entry
    }
    cart.append(new_entry)                                      # Add the new entry to the cart list
    print(f"Added '{item_name}' x{quantity} to the cart.")      # Print confirmation message indicating the item and quantity added to the cart
        
def remove_from_cart(item_name):                                # Function to remove an item from the cart with necessary checks
        for i, present in enumerate(cart):                      # For Loop Iteration through cart to find the item to be removed. Create a variable i and initialize it to 0 to keep track of the index of the item in the cart. Create a variable present to store the current item being checked in the cart.
            if present["item"] == item_name:                    # Check if the item name in the cart matches the item name to be removed
                del cart[i]                                     # If item is found, delete it from the cart using its index
                print(f"Removed '{item_name}' from the cart.")
                return

        print(f"Error: '{item_name}' not found in cart.")          # If item is not found in the cart after checking all items, print an error message indicating that the item was not found in the cart

def update_cart_quantity(item_name, new_quantity):          # Function to update the quantity of an item in the cart with necessary checks
    for updation in cart:                                   # For Loop Iteration through cart to find the item for which quantity needs to be updated. Create a variable updation to store the current item being checked in the cart.  
        if updation["item"] == item_name:                   # Check if the item name in the cart matches the item name for which quantity needs to be updated
            updation["quantity"] = new_quantity             # If item is found, update its quantity to the new quantity provided
            print(f"Set '{item_name}' quantity to {new_quantity}.")  # Print confirmation message indicating the item and its updated quantity
            return
    
    print(f"Error: '{item_name}' not found in cart.")   # If item is not found in the cart after checking all items, print an error message indicating that the item was not found in the cart

# Simulation Steps

print("Step 1: Adding 'Paneer Tikka' x 2")
add_to_cart("Paneer Tikka", 2)              # Call the add_to_cart function to add 'Paneer Tikka' with a quantity of 2 to the cart. This will check if the item exists in the menu, if it is available, and if it is already present in the cart. If all checks pass, it will add the item to the cart and print a confirmation message.
print("Current Cart:", cart, "\n")          # Print the current state of the cart after adding 'Paneer Tikka' to show the user the updated cart contents

print("Step 2: Adding 'Gulab Jamun' x 1")      
add_to_cart("Gulab Jamun", 1)               # Call the add_to_cart function to add 'Gulab Jamun' with a quantity of 1 to the cart. This will perform the same checks as before and update the cart accordingly, followed by printing the current state of the cart to show the user the updated contents after adding 'Gulab Jamun'.
print("Current Cart:", cart, "\n")          # Print the current state of the cart after adding 'Gulab Jamun' to show the user the updated cart contents

print("Step 3: Adding 'Paneer Tikka' x 1")
add_to_cart("Paneer Tikka", 1)             # Call the add_to_cart function to add 'Paneer Tikka' with a quantity of 1 to the cart. Since 'Paneer Tikka' is already present in the cart from Step 1, this will update the existing entry for 'Paneer Tikka' by increasing its quantity from 2 to 3, and print a confirmation message indicating the updated quantity.
print("Current Cart:", cart, "\n")         # Print the current state of the cart after updating 'Paneer Tikka' quantity to show the user the updated cart contents

print("Step 4: Adding 'Mystery Burger' to the cart")
add_to_cart("Mystery Burger", 1)           # Call the add_to_cart function to add 'Mystery Burger' with a quantity of 1 to the cart. Since 'Mystery Burger' does not exist in the menu, this will print an error message indicating that the item does not exist in the menu. 
print("Current Cart:", cart, "\n")         # Print the current state of the cart after attempting to add 'Mystery Burger' to show the user that the cart contents remain unchanged due to the error

print("Step 5: Adding 'Chicken Wings' to the cart")
add_to_cart("Chicken Wings", 1)
print("Current Cart:", cart, "\n")

print("Step 6: Removing 'Gulab Jamun' from the cart")
remove_from_cart("Gulab Jamun")
print("Current Cart:", cart, "\n")

# Final Order Summary

print("======== Order Summary ========")
subtotal = 0                  # Initialize a variable subtotal to keep track of the total cost of items in the cart

for entry in cart:             # For Loop Iteration through cart to calculate the total cost of each item and update the subtotal. Create a variable entry to store the current item being processed in the cart.
    item_total = entry["quantity"] * entry["price"]  # Calculate the total cost of the current item by multiplying its quantity with its price
    subtotal += item_total                           # Update the subtotal by adding the total cost of the current item to the existing subtotal
   
    print(f"{entry['item']:<18} x{entry['quantity']}    ₹{item_total:.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst

print("------------------------------------")
print(f"{'Subtotal:':<20} ₹{subtotal:.2f}")
print(f"{'GST (5%):':<20} ₹{gst:.2f}")
print(f"{'Total Payable:':<20} ₹{total_payable:.2f}")
print("================================")

   
```

```python
# Task 3: Deep Copy and Inventory Tracking

import copy

# Again feeding the inventory in this code block for ease of operation.
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

inventory_backup = copy.deepcopy(inventory)     # Create a deep copy of the inventory to preserve the original state for later verification and simulation purposes

# Example Demonstration of Deep Copy

inventory["Paneer Tikka"]["stock"] = 1000      # Modify the stock of 'Paneer Tikka' in the original inventory to 1000.
print("\n--- Deep Copy  Working Demostration ---\n")
print(f"Inventory (Paneer Tikka): {inventory['Paneer Tikka']['stock']}")
print(f"Backup Inventory (Paneer Tikka): {inventory_backup['Paneer Tikka']['stock']}")

# Restore original state of inventory from its deep copy.

inventory = copy.deepcopy(inventory_backup)

# Final cart from Task 2 was: [{"item": "Paneer Tikka", "quantity": 3, "price": 180.0}]



print("\n--- Processing Order Fulfillment ---")
for cart_item in cart:                  # For Loop Iteration through cart to process each item for order fulfillment. Create a variable cart_item to store the current item being processed in the cart.
    name = cart_item["item"]            # Extract the item name from the current cart item to identify which item is being processed for order fulfillment.
    qty_needed = cart_item["quantity"]  # Extract the quantity needed for the current item from the cart item to determine how many units of the item need to be deducted from the inventory for order fulfillment.
    
    if name in inventory:               # Check if the item being processed for order fulfillment exists in the inventory to ensure that we can update the stock accordingly.
        available = inventory[name]["stock"]   # Get the available stock for the item from the inventory to compare it with the quantity needed for order fulfillment and determine how to update the inventory stock.
        if available >= qty_needed:            # If the available stock is sufficient to fulfill the order, deduct the needed quantity from the inventory stock for that item.
            inventory[name]["stock"] -= qty_needed
        else: 
            print(f" Warning: Insufficient stock for {name}. Deducting from available stock of {available}.")   # If the available stock is insufficient to fulfill the order, print a warning message indicating the item and the available stock, and then set the inventory stock for that item to 0 since all available stock will be used to fulfill as much of the order as possible.
            inventory[name]["stock"] = 0

# Reorder Alert

print("\n--- Reorder Alert ---")
for item, details in inventory.items():  # For Loop Iteration through inventory and get its data as tuples. Create a variable item to store the name(key) of the current item being checked, and a variable details to store the dictionary containing the stock and reorder level information for that item from the inventory.
    if details["stock"] <= details["reorder_level"]:   # Check if the current stock of the item is less than or equal to its reorder level to determine if a reorder alert needs to be generated for that item.
        print(f"Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")
    # No re order alert will be generated since the stock of 'Paneer Tikka' is still above its reorder level after fulfilling the order, and all other items in the inventory remain unchanged with sufficient stock levels.


print("\n--- Updated Inventory After Order Fulfillment ---")
print(inventory)
print("\n--- Inventory Backup (Original State) ---")
print(inventory_backup)
```

```python
# TASK 4 - Daily Sales Log Analysis
# Again feeding the sales data in this code block for ease of operation.

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

highest_revenue = 0
highest_revenue_date = ""
all_order = []
item_count = {}

for date, orders in sales_log.items():      # For Loop Iteration through sales_log to print the sales data for each day. Create a variable date to store the current date being processed, and a variable orders to store the list of orders for that date.
    total_revenue = 0                       # Initialize the total revenue for the current day
    for order in orders:                    # For Loop Iteration through the list of orders for the current date to print the details of each order. Create a variable order to store the current order being processed.
       all_order.append((date, orders))     # Append a tuple containing the date and its corresponding list of orders to the all_order list to keep track of all orders along with their dates for later analysis
       for item in order["items"]:          # For Loop Iteration through the list of items in the current order to count the frequency of each item being ordered. Create a variable item to store the current item being processed in the order.
           item_count[item] = item_count.get(item, 0) + 1 # Update the count of each item being ordered by iterating through the items in the current order and using a dictionary to keep track of the frequency of each item across all orders.
           


       total_revenue += order["total"]          # Update the total revenue by adding the total amount of the current order to the existing total revenue
    print(f" Total revenue for Date {date}: ₹{total_revenue:.2f}")  # Print the total revenue for the current day after processing each order to show the user the cumulative revenue generated from the orders of that day

    if total_revenue > highest_revenue:
        highest_revenue = total_revenue
        highest_revenue_date = date
    
      # Append a tuple containing the date and its corresponding list of orders to the all_order list to keep track of all orders along with their dates for later analysis


print(f"\nHighest Revenue Date: {highest_revenue_date} with ₹{highest_revenue:.2f}")
print(f"Most Ordered Item: {max(item_count, key=item_count.get)} with {max(item_count.values())} orders")

# Add new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

all_order = []  # Reset the all_order list to store the updated orders after adding the new day to the sales log
item_count = {} # Reset the item_count dictionary to count the frequency of items again after adding
print("\n--- Updated Sales Log ---")

for date, orders_new in sales_log.items():      # For Loop Iteration through sales_log to print the sales data for each day. Create a variable date to store the current date being processed, and a variable orders to store the list of orders for that date.
    total_revenue_updated = 0                       # Initialize the total revenue for the current day
    for order in orders_new:                    # For Loop Iteration through the list of orders for the current date to print the details of each order. Create a variable order to store the current order being processed.
       all_order.append((date, order))     # Append a tuple containing the date and its corresponding list of orders to the all_order list to keep track of all orders along with their dates for later analysis
       for item in order["items"]:          # For Loop Iteration through the list of items in the current order to count the frequency of each item being ordered. Create a variable item to store the current item being processed in the order.
           item_count[item] = item_count.get(item, 0) + 1 # Update the count of each item being ordered by iterating through the items in the current order and using a dictionary to keep track of the frequency of each item across all orders.
           

       total_revenue_updated += order["total"]          # Update the total revenue by adding the total amount of the current order to the existing total revenue
    
    print(f" Total revenue for Date {date}: ₹{total_revenue_updated:.2f}")  # Print the total revenue for the current day after processing each order to show the user the cumulative revenue generated from the orders of that day

    if total_revenue_updated > highest_revenue:
        highest_revenue = total_revenue_updated
        highest_revenue_date = date
    
print(f"\nHighest Revenue Date: {highest_revenue_date} with ₹{highest_revenue:.2f}")
print(f"Most Ordered Item: {max(item_count, key=item_count.get)} with {max(item_count.values())} orders")


print("\n--- Numbered Order List ---")
for i, (date, order) in enumerate(all_order, 1):
    items_str = ", ".join(order["items"])
    print(f"{i}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")
```
